```python
from tensorflow.keras.preprocessing.text import one_hot
```

### Sentences

```python
sent=['the glass of milk',
      'the glass of juice',
      'the cup of tea',
      'I am a good boy',
      'I am a good developer',
      'understand the meaning of words',
      'your videos are good']
```

## Define the vocabulary size

```python
voc_size=10000
```

### One Hot Representation

```python
one_hot_repr=[one_hot(words,voc_size) for words in sent]
one_hot_repr
```

## Word Embedding Representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences #we need to make the sentences with the same size
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_lenght=8 #complete with 0 till we get len of 8
embedded_doc=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_lenght)#'pre' add the zeros on the beggining. 'post' is reverse
embedded_doc
```

## Feature representation

```python
dim=10 #google word2vec has 300
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_lenght))#we don't train anything. Embedding layer is internaly using word2vec
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_doc)
```
